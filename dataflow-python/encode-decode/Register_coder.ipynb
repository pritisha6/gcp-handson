This program demonstrates how to manage coders for different data types within Apache Beam. It shows how to check the default coder for a data type (e.g., `int`) and how to register a custom coder for another data type (e.g., `map`).

```python
from apache_beam import coders
#for checking the Encode type of each Data type
coders.registry.get_coder(int)

#For changing the encode type of each Data type in Apache beam
coders.registry.register_coder(map,coders.StrUtf8Coder)
coders.registry.get_coder(map)
```
