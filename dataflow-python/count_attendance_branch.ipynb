Count attendance of each employee of both HR and Accounts Department.
Input File has schema as below:


```
EmpId,name,DeptId,DeptName,Date
241316NX,Kumiko,10,Accounts,1-01-2019
796656IE,Gaston,10,Accounts,1-01-2019
331593PS,Beryl,20,HR,1-01-2019
560447WH,Olga,20,HR,1-01-2019
```

Output is expected as


```
('Accounts Kumiko',1)
('Accounts Gaston',1)
('HR Beryl',1)
('HR Olga',1)
```

SQL query is -



```
select distinct Emp_Name,dept_name,count(1) as attendance
FROM `pritiproject.Apache_beam_tables.dept_data` 
where Dept_Name='Accounts' or Dept_Name='HR'
group by 1,2
```

````python
import apache_beam as beam

p1 = beam.Pipeline()

input_collection = (
    p1
    | "Read from a file" >> beam.io.ReadFromText('/content/dept_data.txt')
    | "seperate each column value" >> beam.Map(lambda record: record.split(','))
    # | "Write to file" >> beam.io.WriteToText('/content/output.txt')
)
accounts_collection = (
    input_collection
    | "Filter accounts department" >> beam.Filter(lambda record: record[3] == 'Accounts')
    | "Pair accounts employee with 1" >> beam.Map(lambda record: ("Accounts "+record[1], 1))
    | "Group and sum accounts" >> beam.CombinePerKey(sum)
    | "Write accounts result" >> beam.io.WriteToText('/content/output_accounts.txt')
)
hr_collection = (
    input_collection
    | "Filter hr department" >> beam.Filter(lambda record: record[3] == 'HR')
    | "Pair hr employee with 1" >> beam.Map(lambda record: ("HR "+record[1], 1))
    | "Group and sum hr" >> beam.CombinePerKey(sum)
    | "Write hr result" >> beam.io.WriteToText('/content/output_hr.txt')
)

p1.run()

!{('head -n 20 /content/output*')}
````
