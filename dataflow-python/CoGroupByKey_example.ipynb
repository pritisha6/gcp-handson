```python
import apache_beam as beam
p1=beam.Pipeline()

def reTuple(element):
  record = element.strip(',')
  return (record[0],record[1:])

dept_data = (p1
             |'read dept_data'>>beam.io.ReadFromText('/content/dept_data.txt')
             |'split dept data'>>beam.Map(reTuple)
             )
            #  |'write dept_data'>>beam.io.WriteToText('/content/output-dept'))

loc_data = (p1
            |'read location_data'>>beam.io.ReadFromText('/content/location.txt')
            |'split location data'>>beam.Map(reTuple)
            )
            # |'write location_data'>>beam.io.WriteToText('/content/output-loc'))
co_group_output = ({'dept_data':dept_data,'loc_data':loc_data}
|beam.CoGroupByKey()
|'final output'>>beam.io.WriteToText('/content/output-final')
)
p1.run()
!{('head -n 5 /content/output-*')}
```
